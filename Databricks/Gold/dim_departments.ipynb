```python
%sql
CREATE TABLE IF NOT EXISTS gold.dim_department
(
Dept_Id string,
SRC_Dept_Id string,
Name string,
datasource string
)
```

```python
%sql 
truncate TABLE gold.dim_department ;


insert into gold.dim_department
select 
distinct
Dept_Id ,
Src_DeptId ,
Name ,
datasource 
 from silver.departments
 where is_quarantined=false
```
